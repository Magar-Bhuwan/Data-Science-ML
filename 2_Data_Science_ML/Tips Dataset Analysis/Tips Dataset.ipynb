```python
grouped_data = tips_data.groupby(['sex', 'smoker'])
```

```python
grouped_data[['tip', 'total_bill']].mean()
```

Here we can clearly see the data of tip and total bill is much more spend male than female.

```python
tips_data[['tip', 'total_bill']].corr()
```

#### Use correlation when you want to measure the relationship between two numeric variables.

It helps answer questions like:

- “When one variable increases, does another also increase?”
- “Are these variables related?”
- “How strongly are they connected?”

--> As bill increases, tip also increases.

---> That creates positive correlation.

So customers with larger bills tend to give larger tips.
